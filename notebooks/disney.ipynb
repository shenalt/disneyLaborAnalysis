# Import pandas

```python
import pandas as pd
#pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.reset_option('display.max_rows')
#pd.reset_option('display.max_columns')
```

# Load in the dataframes

```python
df = pd.read_csv("data/laborSummaryData.csv") 
df = df[df["annual_average_employees"] > 0]
df["All Types of Injuries"] = df["total_injuries"] + df["total_skin_disorders"] + df["total_respiratory_conditions"] + df["total_poisonings"] + df["total_hearing_loss"] + df["total_other_illnesses"]
df
```

# Sort the dataframe by the companies with the largest amount of injuries reported

```python
df.sort_values(by="All Types of Injuries", ascending=False)
```

# Disney looks to have the most occurrences and it is listed mutiple times so I am going to isolate it

```python
# Make the establishment_name column all lowercase first
df['establishment_name'] = df['establishment_name'].str.lower()

# Filter for only Disney entries
disney = df[df["establishment_name"].str.contains('disney')]

# Save as csv to output folder
disney.to_csv("output/disney.csv", index=False)

disney
```

# How many injuries across all Disney locations were there in 2024?

```python
disney["All Types of Injuries"].sum()
```

# How many average employees are there across all the locations in 2024?

```python
disney["annual_average_employees"].sum()
```

# Rate of injury across Disney locations based on average employees

```python
disney_injury_rate = disney["All Types of Injuries"].sum() / disney["annual_average_employees"].sum()
disney_injury_rate
```

# The high number of injuries is notable but the rate of injury is low as Disney employs many people. Curious to see rates across the dataset

```python
# Measuring just total injuries
df["Injuries per Average Employee"] = df["total_injuries"] / df["annual_average_employees"]
df.sort_values(by="Injuries per Average Employee", ascending=False).head(10)
```

```python
# Measuring all types of injuries rate
df["All Types of Injuries per Average Employee"] = df["All Types of Injuries"] / df["annual_average_employees"]
df.sort_values(by="All Types of Injuries per Average Employee", ascending=False).head(15)
```

```python
all_inj_sector_summ = df.groupby('sector')[['All Types of Injuries', 'annual_average_employees']].sum().reset_index()
all_inj_sector_summ["Injury Rate"] = all_inj_sector_summ['All Types of Injuries'] / all_inj_sector_summ['annual_average_employees']
all_inj_sector_summ.to_csv("output/all_inj_sector_summ.csv", index=False)
all_inj_sector_summ.sort_values(by='Injury Rate', ascending=False)
```

```python
all_death_sector_summ = df.groupby('sector')[['total_deaths', 'annual_average_employees']].sum().reset_index()
all_death_sector_summ["Death Rate"] = all_death_sector_summ['total_deaths'] / all_death_sector_summ['annual_average_employees']
all_death_sector_summ.to_csv("output/all_death_sector_summ.csv", index=False)
all_death_sector_summ.sort_values(by='Death Rate', ascending=False)
```
